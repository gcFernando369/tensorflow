```python
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow import keras
import numpy as np
from matplotlib import pyplot as plt
```

# Learning sin function

```python
x=np.arange(-1,1.05,0.05)
y=np.sin(x*np.pi)
```

```python
plt.plot(x,y)
plt.title('sin(x) function')
plt.show()
```

# Build model

```python
model=keras.Sequential([
    layers.Dense(32,activation=tf.nn.relu, input_shape=[1]),
    layers.Dense(32,activation=tf.nn.relu),
    layers.Dense(1)
])

model.compile(
    loss='mse',
    optimizer=keras.optimizers.Adam(lr=0.001),
    metrics=['mae']
)
```

# Train model

```python
model.fit(x,y,epochs=1000,verbose=0)
```

# Watch result

```python
result=model.predict(x)
plt.plot(x,result)
plt.title('prediction')
plt.show()
```

# Learning x^3 function

```python
x=np.arange(-1,1.05,0.05)
y=x**3
```

```python
plt.plot(x,y)
plt.title('x^3function')
plt.show()
```

# Train the same model

```python
model.fit(x,y,epochs=1000,verbose=0)
```

# Watch result

```python
result=model.predict(x)
plt.plot(x,result)
plt.title('prediction')
plt.show()
```
